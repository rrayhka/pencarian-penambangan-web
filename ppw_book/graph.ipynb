```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8, 9])  # Add nodes 1 to 5

# Add edges
G.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 3), (3,4), (4,5), (4,6), (5,7), (5,8), (6,7), (6,8), (7,8), (9,7),(6,5)]) 
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=700, font_size=16, font_color='black', font_weight='bold')
plt.title(f'Random Graph with 9 Nodes')
plt.show()

print("Degree Centrality:")
for node, centrality in degree_centrality.items():
    print(f"Node {node}: {centrality:.2f}")

print("\nCloseness Centrality:")
for node, centrality in closeness_centrality.items():
    print(f"Node {node}: {centrality:.2f}")
```
